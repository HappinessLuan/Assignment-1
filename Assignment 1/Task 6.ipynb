```python
import random

# 3-digit code: each number between 0 and 9
code1 = f'{random.randint(0, 9)}{random.randint(0, 9)}{random.randint(0, 9)}'

# 4-digit code: each number between 1 and 6
code2 = f'{random.randint(1, 6)}{random.randint(1, 6)}{random.randint(1, 6)}{random.randint(1, 6)}'

print(f'3-digit code (0-9): {code1}')
print(f'4-digit code (1-6): {code2}')
```
